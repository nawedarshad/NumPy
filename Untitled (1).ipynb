```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open("OpenCV/1.jpg")
#opening Image.
```

```python
type(pic)
```

```python
pic_array = np.asarray(pic)
#storing pic as an araay
```

```python
pic_array.shape
#prints the shape of array
```

```python
plt.imshow(pic_array)
#shows image of array
```

```python
plt.imshow(pic_array[:,:,1])
#only gree Channels
```

```python
#only Red Channels
plt.imshow(pic_array[:,:,0])
```

```python
#only Blue Channel
plt.imshow(pic_array[:,:,2])
```

```python
pic_array[:,:,1]
#prints Array
```
